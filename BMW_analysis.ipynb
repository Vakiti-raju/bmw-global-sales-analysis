```python
#Import pandas matplotlib and seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('bmw_global_sales_2018_2025.csv')
df.head()
```

```python
Year_sales = df.groupby('Year')['Units_Sold'].sum()
Year_sales.plot(kind= 'line', color = 'Green')
plt.title("YEARLY SALES TREND")
plt.show()
```

```python
Regional_sales = df.groupby('Region')['Units_Sold'].sum()
Regional_sales.plot(kind = 'bar')
plt.ylabel('units sold')
plt.title('REGIONAL SALES TREND')
plt.show()
```

```python
Model_values = df.groupby('Model')['Units_Sold'].sum().sort_values(ascending=False)
Model_values.head(10).plot(kind = 'bar')
plt.title('Top models')
plt.show()
```

```python
plt.scatter(df["Units_Sold"],df["Revenue_EUR"])
plt.xlabel('Units sold')
plt.ylabel('Revenue')
plt.show()
```

```python
average_trend = df.groupby('Fuel_Price_Index')['BEV_Share'].mean()
average_trend.plot()
plt.title('Fuel Price vs BEV adoption')
plt.xlabel('Fuel index')
plt.ylabel('BEV share')
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True),
annot=True)
plt.show()
```

```python
Regional_share= df.groupby('Region')['Units_Sold'].sum()
Regional_share.plot(kind= 'pie', autopct= '%1.1f%%')
plt.title('Market share analysis')
plt.ylabel("")
plt.show()
```

```python
model_revenue = df.groupby('Model')['Revenue_EUR'].sum().sort_values(ascending=False)
model_revenue.head(10).plot(kind = 'bar')
plt.title("Top revenue generating models")
plt.xlabel('Models')
plt.ylabel('Revenue (in million)')
plt.show()
```

```python
BEV_growth =df.groupby('Year')['BEV_Share'].mean()
BEV_growth.plot(kind = 'bar', color = 'green')
plt.xlabel('YEAR')
plt.ylabel('BEV_SHARE')
plt.show()
```
